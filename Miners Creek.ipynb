```python
import numpy as np
import pandas as pd
```

```python
miners = pd.read_csv('miners_obs.csv')
```

```python
miners.head()
```

```python
#change name of column
miners.columns = ['comid', 'year', 'month', 'value']
```

```python
miners.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#line plot for multi-year mean stream flow for Miners Creek
ax = sns.relplot(x="month", y="value", hue="year", kind="line", data=miners)

```

```python
#create a facet grid for Miners Creek stream flow
sns.relplot(x="month", y="value", hue="year",
            col="year", col_wrap=4, height=6, aspect=.75, linewidth=2.5, kind='line', data=miners);


ax.set_titles('Mean Observed Annual Stream Flow for Miners Creek')
```
