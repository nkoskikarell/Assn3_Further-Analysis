```python
import numpy as np
import pandas as pd
```

```python
miners = pd.read_csv('miners_2000-2015.csv')
```

```python
miners.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#line plot for multi-year mean stream flow for Miners Creek
ax = sns.relplot(x="month", y="value", hue="year", kind="line", data=miners)
```

```python
#line plot for multi-year mean stream flow for Miners Creek
ax = sns.relplot(x="month", y="value", hue="year", style="statistic", kind="line", legend = "full", data=miners, palette="Blues")
```

```python
#line plot for multi-year mean stream flow for Miners Creek
ax = sns.relplot(x="month", y="value", hue="year", style="statistic", kind="line", legend = "full", data=miners, palette="Blues")
```

```python


#create a facet grid for Miners Creek stream flow
sns.relplot(x="month", y="value", hue="year", style = "statistic",
            col="year", col_wrap=4, height=6, aspect=.75, linewidth=2.5, kind='line', legend = "full", data=miners);
```
