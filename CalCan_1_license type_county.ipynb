```python
# plot top 10 most frequently awarded license types
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
top10licensetypes = cultlic.loc[(cultlic['license'].str.contains('provadu_sm_mix_tier')) |
    (cultlic['license'].str.contains('provadu_sm_out')) |
    (cultlic['license'].str.contains('provmed_sm_mix_tier1'))|
    (cultlic['license'].str.contains("provmed_sm_out")) |
    (cultlic['license'].str.contains("anadu_sm_out")) |
    (cultlic['license'].str.contains("provadu_sm_mix_tier2")) |
    (cultlic['license'].str.contains("provadu_sm_ind")) |
    (cultlic['license'].str.contains("provadu_spec_out")) |
    (cultlic['license'].str.contains("provadu_spec_mix_tier1")) &
    (cultlic['license'].str.contains("provadu_nurs"))]
  
```

```python

top10licensetypes['license'].value_counts().plot.bar(edgecolor = 'k')
plt.title('Top 10 Cultivation License Types Awarded in California, 2019')
plt.xlabel('License Type')
plt.ylabel('Number of Licenses')
```
